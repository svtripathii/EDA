## Student Performance 
Student Performance Data was obtained in a survey of students' math course in secondary school.
It consists of 33 Column
Dataset Contains Features like

    school ID
    gender

    size of family
    Father education
    Mother education
    Occupation of Father and Mother
    Family Relation
    Health
